# [A simple example](https://matplotlib.org/stable/users/explain/quick_start.html#a-simple-example)

Matplotlib contém vários objetos e funções.  Seu objeto mais fundamental chama-se  **Figure** correspondendo a uma área onde MatPlotLib desenha o gráfico desejado.

Cada Figure pode conter um ou mais **Axes** que são objetos que correspondem a um sistema de eixos cartesianos, de coordenadas polares ou esféricas.  Os objetos Axes possuem um método **plot** que recebe ao menos dois objetos Python sequenciais para servirem como dados a serem plotados na Figure.

O jeito mais simples de criar uma figura é usar a função subplots() do módulo pyplot e, então, usar o método plot do objeto Axes para desenhar.  Note que esta função retorna uma tupla contendo um objeto Figure e um conjunto de eixos neste objeto.
Veja o código a seguir:




```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4], [1,4,2,3])
```

## Axes

Os objetos [Axes](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.html#matplotlib.axes.Axes) são os eixos do sistema de coordenadas do objeto `Figure`.  `Axes` são um tipo de `Artist` (que é uma classe abstrata) como veremos a seguir.  Os `Axes` são o sistema de eixos ao passo que os `Axis` são os eixos propriamente ditos.  Os `Axis` fornecem as marcações no eixo e a escala dos dados.  Os `Axes` possuem também os atributos `title`, `x-label` e `y-label` que são setados com os métodos `set_itle`, `set_x-label` e `set-y_label`, respectivamente.

Para usar a API orientada a objeto do MatPlotLib, a classe `Axes` são os "entry-points" principais.


## Axis

Os objetos [Axis](https://matplotlib.org/stable/api/axis_api.html#matplotlib.axis.Axis) dão a escala e os limites do sistema de eixos e as marcações de escala no sistema de eixos.

A localização das marcações (ticks) é determinada por classes no módulo [ticker](https://matplotlib.org/stable/api/ticker_api.html#matplotlib.ticker.Formatter).  Este módulo é composto por uma classe abstrata chamada `Locator` e a formatação dos strings que compõem os ticklabels são determinadas pela classe `Formater` do mesmo módulo.

A subclasse `MultipleLocator` da classe abstrata `Locator` é inicializada com uma base (que é a distância entre dois ticks) e  automaticamente preenche o eixo do início até o limite final.

## Artist

É uma classe abstrata que reúne tudo o que é visível dentro de um Figure, inclusive o próprio Figure.
